## Feature Scaling

Feature Scaling is a method to scale numeric features in the same scale or range (like -1 to 1, 0 to 1).
This last step involved in data processing and before ML model training.
it also called as data normalization.
we apply it on independent variables.
we fit feature scaling with train data and transform on train and test data.


    

Why feature scaling?

The scale of raw features ois different according to its units.
Machine leaning algorithams can't understand features units, understand only numbers.
Ex: if hight 140cm and 8.2 feet
ML algorithams understand numbers then 140 > 8.2
so this is the drawback

Standarization?
standarization rescale the feature such as mean = 0 and standard deviation = 1

The result of stadarization is z called as z-score normalization
if data follow normal distrubution(gaussian distribution)--- optional
if the original distribution is normal, then standard distribution will be also normal.


## Normalization
normalization rescale the feature in fixed range between 0 to 1
normalization also called as min-max scaling.
if data doesn't follow nornal distribution.

```python
# Normalization using Min-Max Scaling
df_normalized = (df3 - df3.min()) / (df3.max() - df3.min())
```

```python
df_normalized.head()
```

```python
df3.min()
```

```python
df3.max()
```

```python
# Normalize only columns 'age' and 'parch'
df_specific_normalized = df3.copy()
df_specific_normalized[['age', 'parch']] = (df3[['age', 'parch']] - df3[['age', 'parch']].min()) / (df3[['age', 'parch']].max() - df3[['age', 'parch']].min())
```

```python
df_specific_normalized
```

# Standardization using Z-score normalization

```python

df_standardized = (df3 - df3.mean()) / df3.std()
```

```python
df_standardized
```

```python
df_standardized.describe().round(2)
```

# Original DataFrame: The values in each feature are in their original scales.

# Values less than 0 indicate that they are below the mean.
# Values greater than 0 indicate that they are above the mean.
