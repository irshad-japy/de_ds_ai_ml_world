```python
df_copy = df.copy()

# List of columns with missing values
num_vars_miss = ['age', 'gender', 'device_type', 'ad_position', 'browsing_history', 'time_of_day']

# Loop through the missing variables and the unique classes of 'click'
for miss_var in num_vars_miss:
    for var_class in df['click'].unique():
        # Check if the column is numeric
        if pd.api.types.is_numeric_dtype(df[miss_var]):
            # If numeric, calculate the mean for the current variable and class
            mean_value = df[df['click'] == var_class][miss_var].mean()
            # Fill NaN values with the mean
            df_copy.loc[df_copy['click'] == var_class, miss_var] = df_copy.loc[df_copy['click'] == var_class, miss_var].fillna(mean_value)
        else:
            # If non-numeric, calculate the mode for the current variable and class
            mode_series = df[df['click'] == var_class][miss_var].mode()
            if not mode_series.empty:
                mode_value = mode_series[0]  # Get the first mode value
                # Fill NaN values with the mode
                df_copy.loc[df_copy['click'] == var_class, miss_var] = df_copy.loc[df_copy['click'] == var_class, miss_var].fillna(mode_value)

# Display the updated DataFrame
df_copy
```
